# Compare the Similarity of Artists by Song Similarity

#### Our goal is to compare the similarity of songs between artists, and use this as a metric of how similar artists are to one another. In this code, we compare 3 artists chosen randomly (Foo Fighters, Snoop Dogg, and Selena Gomez), and found the song that is most similar to each song. This will make more sense at the end of the notebook, but the conclusion was very interesting.

```python
# Import libraries and upload OHCO (by line) #
import pandas as pd
import numpy as np
import os
os.chdir('/Users/nickbruno/Documents/spring_2019/DS5559/project/code')

df = pd.read_csv('artist_song_verse_line_OHCO_df.csv') # includes corpus with no duplicate song names
```

```python
# Choose songs from 3 artists in our corpus to compare #
artist_id_list = df.artist_id.unique().tolist()
import random
random.seed(100) # Sets the seed so that we can run the code again with the same outcome
sub_artist = random.sample(artist_id_list, k=3)
    # Takes the artist_id of 3 random artists from our 643 total artists
sub_artist
```

```python
# Subset to take only the songs by the 3 artists #
subset_df = df[df.artist_id.isin(sub_artist)]
```

```python
# Reset the index #
subset_df = subset_df.set_index(['artist_id','song_id','verse_num','line_num'])
```

```python
# Create a bag of tokens #
TOKEN_PAT = r'(\W+)'
token_subset_df = subset_df.Line.str.split(TOKEN_PAT, expand=True)\
    .stack()\
    .to_frame()\
    .rename(columns={0:'token_str'})
token_subset_df.index.names = ['artist_id','song_id','verse_num','line_num','token_num']
```

```python
# Remove numbers and punctuation #
token_subset_df['punc'] = token_subset_df.token_str.str.match(r'^[\W_]*$').astype('int')
token_subset_df['num'] = token_subset_df.token_str.str.match(r'\d').astype('int')
```

```python
# Create subsets where punc and num == 0
no_punc = token_subset_df[token_subset_df.punc == 0]
no_num = no_punc[no_punc.num == 0]
```

```python
no_num.token_str = no_num.token_str.str.lower()
```

```python
# Remove stopwords #
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
final_token_df = no_num[-no_num.token_str.isin(stop)] # removes stopwords from our bag of words
```

```python
final_token_df = final_token_df.drop(columns=['punc','num']) # drops unnecessary columns
```

```python
df = final_token_df
```

```python
df.head()
```

#### Now we have a bag of words that excludes spaces, punctuation, and stopwords. We can now go into the heart of our analysis

```python
# Import important nltk packages
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('tagsets')
nltk.download('wordnet')
```

```python
# Stem each word #
stemmer = nltk.stem.porter.PorterStemmer()
df['port_stem'] = df.token_str.apply(lambda x: stemmer.stem(x))
```

```python
token_counts = df.port_stem.str.lower().value_counts()
```

```python
# Create a vocab table for our subset of 3 artists #
vocab = df.port_stem.value_counts().to_frame()\
    .reset_index()\
    .rename(columns={'index':'port_stem', 'port_stem':'n'})
vocab = vocab.sort_values('port_stem').reset_index()
vocab.index.name = 'term_id'
vocab = vocab.drop('index', 1)
```

```python
# Organize our dataframes #
K = df
V = vocab
```

```python
# Add a unique id for each stem word #
K['term_id'] = K['port_stem'].map(V.reset_index().set_index('port_stem').term_id).fillna(-1).astype('int')
```

```python
OHCO = ['artist_id', 'song_id', 'verse_num', 'line_num','token_num']
SONG=OHCO[1:2] # just looks at 'song_id'
```

```python
# Create a bag of words per song #
BOW = K.groupby(SONG+['term_id'])['term_id'].count()
BOW.head()
```

```python
# Create a document-term matrix
DTM = BOW.unstack().fillna(0)
DTM
```

#### Create the TFIDF table

```python
# Now creating the TFIDF table #
alpha = .000001 # We introduce an arbitrary smoothing value
alpha_sum = alpha * V.shape[0]
TF = DTM.apply(lambda x: (x + alpha) / (x.sum() + alpha_sum), axis=1)
```

```python
N_docs = DTM.shape[0]
V['df'] = DTM[DTM > 0].count()
TFIDF = TF * np.log2(N_docs / V['df'])
```

```python
TFIDF.head() # now we have the TFIDF table
```

```python
THM = -(TF * np.log2(TF))
TFTH = TF.apply(lambda x: x * THM.sum(), 1)
```

```python
# Add all of this TFIDF information to the original 'Vocab' table #
V['tf_sum'] = TF.sum()
V['tf_mean'] = TF.mean()
V['tf_max'] = TF.max()
V['tfidf_sum'] = TFIDF.sum()
V['tfidf_mean'] = TFIDF.mean()
V['tfidf_max'] = TFIDF.max()
V['tfth_sum'] = TFTH.sum()
V['tfth_mean'] = TFTH.mean()
V['tfth_max'] = TFTH.max()
V['th_sum'] = THM.sum()
V['th_mean'] = THM.mean()
V['th_max'] = THM.max()
```

```python
# Find significant words #
TOPS = pd.DataFrame(index=range(10))
for m in ['tf','tfidf','th']:
    for n in ['mean','max']:
        key = '{}_{}'.format(m,n)
        TOPS[key] = V.sort_values(key, ascending=False).port_stem.head(10).tolist()
```

```python
# Define a function to get songs from our subset tokens table #
def get_songs(tokens, div_names, doc_str = 'term_id', sep='', flatten=False, 
             index_only=False):
    
    if not index_only:
        docs = tokens.groupby(div_names)[doc_str]\
          .apply(lambda x: x.str.cat(sep=sep))
        docs.columns = ['doc_content']
    else:
        docs = tokens.groupby(div_names)[doc_str].apply(lambda x: x.tolist())

    if flatten:
        docs = docs.reset_index().drop(div_names, 1)
    
    return docs
```

```python
# Create a dataframe that includes the term_count and term-frequency #
D = get_songs(K, SONG, 'port_stem', sep=" ")\
    .to_frame().rename(columns={'port_stem':'song_str'})
```

```python
D['term_count'] = DTM.sum(1).astype('int').to_frame().rename(columns={0:'term_count'})
D['tf'] = D.term_count / D.term_count.sum()
```

```python
D.head()
```

```python
song_ids = D.index.tolist()
pairs = [(i,j) for i in song_ids for j in song_ids if j > i]
```

```python
# Create a data frame to compare the similarity between a pair of songs #
P = pd.DataFrame(pairs).reset_index(drop=True).set_index([0,1])
P.index.names = ['song_x','song_y']
```

```python
P.head()
```

```python
# Define a function that calculates Euclidean distance #
def euclidean(row):
    D1 = TFIDF.loc[row.name[0]]
    D2 = TFIDF.loc[row.name[1]]
    x = (D1 - D2)**2
    y = x.sum() 
    z = np.sqrt(y)
    return z
```

```python
# Find euclidean distance between songs
P['euclidean'] = 0
P['euclidean'] = P.apply(euclidean, 1)
```

```python
# Create a dataframe (PME) that shows the similarity between songs in terms of Euclidean distance #
P['euclidean_sim'] = 1 / (P.euclidean + 1)
PME = P.euclidean.unstack().fillna(0)
PME.head()
```

We notice that the similarities between doc_x and doc_y in the beginning of the table shows a lot of zeros. This is an issue that we correct later in the code.

```python
# Define a function that finds cosine similarity #
def cosine(row):
    D1 = TFIDF.loc[row.name[0]]
    D2 = TFIDF.loc[row.name[1]]
    x = D1 * D2
    y = x.sum()
    
#     a = np.sqrt(D1.sum()**2)     # <-- misplaced exponent!
#     b = np.sqrt(D2.sum()**2)     # <-- misplaced exponent!
#     c = np.sqrt(a) * np.sqrt(b)  # <-- extra square roots!

    a = np.sqrt((D1**2).sum())
    b = np.sqrt((D2**2).sum())
    c = a * b
    
    z = y / c
    return z
```

```python
# Apply our cosine function to compute cosine similarity between songs
P['cosine'] = P.apply(cosine, 1)
P.head()
```

```python
# Create a table that compares these similarities like a confusion matrix
PMC = P.cosine.unstack().fillna(0)
PMC.head()
P.cosine.sort_values(ascending=True).head(50).plot(kind='barh', figsize=(5,10))
    # Shows the 50 most similar songs within our subset
```

```python
# Create a dataframe that unstacks PMC #
pmc_df = PMC.unstack()
pmc_df = pd.DataFrame(pmc_df)
pmc_df.doc_y = pmc_df.index
```

```python
pmc_df = pmc_df.reset_index(level=['song_y', 'song_x'])
pmc_df = pmc_df.rename(columns={0:'Similarity'})
```

```python
pmc_df.head()
```

Again, we see an issue with many of the similarities of the same artist with a similarity = 0. We will now fix this by making a separate dataframe that mimicks 'pmc_df', but switches the 'song_y' and 'song_x' columns. This keeps the integrity of the similarities while finding the true highest similarity for each song. For example, let's say we have two songs with song_id = 1 and song_id = 2, with a similarity = 0.75. The issue with our current dataframe is that if 'song_y' is 1 and 'song_x' is 2, the 'Similarity' column would equal 0.75. However, if song 2 becomes 'song_y' and song 1 becomes 'song_x', the dataframe would have a value of 0 for the 'Similarity' column. By creating a new dataframe that recognizes that song 2's highest similarity is 0.75 with song 1 by flipping the 'song_y' and 'song_x' labels, this will ensure that none of the songs have their highest similarity between another song equal to 0. This often leads to duplicates within the dataframe, but we will also take care of that later in the code.

```python
pmc_df.shape[0] # number of similaritiese to compare
```

```python
len(pmc_df.song_y.unique()) # 163 different songs
```

```python
# Create a new dataframe that mimics our 'pmc_df' dataframe, but switches the doc_y and doc_x #
pmc_df_new = pmc_df
pmc_df_new.columns = ['song_x','song_y','Similarity']
```

```python
pmc_df_new.head()
```

```python
pmc_df_new.shape[0] # maintains the same shape
```

```python
# Merge our original 'pmc_df' dataframe with 'pmc_df_new' to create an accurate similarity table #
pmc_df_merge = pmc_df.append(pmc_df_new)
pmc_df_merge.shape[0] # has twice as many rows as expected
```

```python
# Create a dataframe that finds the highest similarity scores for each 'song_y' #
highest_sim = pmc_df_merge.groupby('song_y')[['Similarity']].max().sort_values('Similarity',ascending=False)
highest_sim = highest_sim.reset_index(['song_y'])
highest_sim.tail()
```

This creates a dataframe that contains the highest similarity between a song (song_y) and its associated Similarity score. What we can now do is inner join this dataframe with our previous 'pmc_df_merge' to find the highest Similarity score of each song in our subsetted dataframe containing the songs from 3 artists. As we can see, there are not values of 0 in the highest_sim table, meaning that our solution to the earlier problem has been resolved.

```python
# Inner join 'highest_dim' with 'pmc_df_merge' #
sim_merge = pd.merge(highest_sim, pmc_df_merge, how='inner', on=['song_y','Similarity'])
```

```python
# drop duplicate rows #
sim_merge_no_dup = sim_merge.drop_duplicates()
sim_merge_no_dup.shape[0] # Has the same number of rows as original songs (163)
```

```python
# Show the correct similarity table #
sim_merge_no_dup.head()
```
